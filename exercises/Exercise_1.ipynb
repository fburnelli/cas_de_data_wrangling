### Facultative: json deep dive

````python
pretty_json = json.dumps(json.loads(df.head(1).to_json(orient='records')), indent=4)
print(pretty_json)
````

````python
pretty_json = json.dumps(json.loads(df.head(1).to_json(orient='index')), indent=4)
print(pretty_json)
````

````python
pretty_json = json.dumps(json.loads(df.head(1).to_json(orient='columns')), indent=4)
print(pretty_json)
````
